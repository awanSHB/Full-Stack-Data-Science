```python
## similarity and dissimilarity matrix

from scipy.spatial.distance import pdist, squareform

# Select two attributes for comparison
# a1 = 'Rating'
# a2 = 'avg_salary'

a1 = 'spark'
a2 = 'aws'

# Compute Euclidean distance between the two attributes
dissimilarity_matrix = pdist(df1[[a1, a2]], metric='euclidean')
dissimilarity_matrix = squareform(dissimilarity_matrix)

print("Dissimilarity Matrix:")
dissimilarity_matrix
```
