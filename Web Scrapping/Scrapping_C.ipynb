```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial.distance import pdist, squareform

# Numeric columns for Min-Max scaling
numeric_cols = ['rating', 'price', 'original_price']

# Categorical columns for Jaccard similarity
categorical_cols = ['brand', 'ram', 'rom', 'expandable_size']

# Drop rows with missing values in categorical columns
dfc.dropna(subset=categorical_cols, inplace=True)

# Normalize numeric attributes
scaler = MinMaxScaler()
dfc[numeric_cols] = scaler.fit_transform(dfc[numeric_cols])

# Compute Euclidean distance for numeric attributes
numeric_distances = pdist(dfc[numeric_cols], metric='euclidean')
numeric_similarity_matrix = 1 / (1 + squareform(numeric_distances))

# Compute Jaccard similarity for categorical attributes
jaccard_similarities = pd.DataFrame(index=dfc.index, columns=dfc.index)
for i in range(len(dfc)):
    for j in range(len(dfc)):
        if i != j:
            common_elements = sum(dfc[categorical_cols].iloc[i] == dfc[categorical_cols].iloc[j])
            total_elements = len(categorical_cols)
            jaccard_similarities.iloc[i, j] = common_elements / total_elements

# Combine similarity matrices
similarity_matrix = 0.5 * (numeric_similarity_matrix + jaccard_similarities)

# Dissimilarity matrix
dissimilarity_matrix = 1 - similarity_matrix

print("Similarity Matrix:")
print(similarity_matrix)

print("\nDissimilarity Matrix:")
print(dissimilarity_matrix)
```
