```python
driver.get("https://www.flipkart.com/search?q=mobile+phones&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_4_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobile+phones&requestId=e625b409-ca2a-456a-b53c-0fdb7618b658&as-backfill=on")
```

```python
content = driver.page_source
soup = BeautifulSoup(content, 'html.parser')
```

```python
brand = []
model = []
rating = []
reviews = []
memory = []
size = []
camera = []
battery =  []
processor = []
warranty = []
price = []
original_price = []
dicounted_percentage = []
```

```python
for k in soup.find_all('div', {'class':'_2kHMtA'}):
    t = k.find('div', class_='_4rR01T').text.strip()
    model.append(t)

    r = k.find('div', class_='_3LWZlK').text.strip()
    rating.append(r)

    rc = k.find('span', class_='_2_R_DZ').text
    reviews.append(rc)

    sp = k.find_all('li', class_='rgWa7D')
    m = sp[0].text.strip() if len(sp) > 0 else ''
    s = sp[1].text.strip() if len(sp) > 1 else ''
    c = sp[2].text.strip() if len(sp) > 2 else ''
    b = sp[3].text.strip() if len(sp) > 3 else ''
    p = sp[4].text.strip() if len(sp) > 4 else ''
    w = sp[5].text.strip() if len(sp) > 5 else 'No Warranty'
    
    memory.append(m)
    size.append(s)
    camera.append(c)
    battery.append(b)
    processor.append(p)
    warranty.append(w)

    pr = k.find('div', class_='_30jeq3 _1_WHN1').text.strip()
    price.append(pr)

    op = k.find('div', class_='_3I9_wc _27UcVY').text.strip()
    original_price.append(op)

    dp = k.find('div', class_='_3Ay6Sb').text.strip()
    dicounted_percentage.append(dp)
```
