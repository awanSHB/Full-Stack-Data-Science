```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/mobile-phones-store?fm=neo%2Fmerchandising&iid=M_9cb46539-c90f-42e8-a7d4-eec1e2cb7f7b_1_372UD5BXDFYS_MC.ZRQ4DKH28K8J&otracker=hp_rich_navigation_2_1.navigationCard.RICH_NAVIGATION_Mobiles_ZRQ4DKH28K8J&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_2_L0_view-all&cid=ZRQ4DKH28K8J')
```

```python
content = driver.page_source
soup = BeautifulSoup(content, 'html.parser')
```

```python
model = []
picture = []
URL = []
rating = []
new_price = []
original_price = []
discount_percentage = []
```

```python
for k in soup.find_all('div', {'class': '_2n7i6c'}):
    t = k.find('a', class_='_2rpwqI')
    mn = t.get('title')
    model.append(mn)

    pic = t.find('img').get('src')
    picture.append(pic)

    url = t.get('href')
    URL.append(url)

    r = k.find('div', class_='_3LWZlK')
    rat = r.text.strip()
    rating.append(rat)

    # p = k.find('div', class_='_25b18c')
    # pri = p.find('div', class_='_8VNy32')
    # pr = p.find('div', class_='_30jeq3').text.strip()
    # # op = p.find('div', class_='_3I9_wc').text.strip()
    # dp = p.find('div', class_='_3Ay6Sb').text.strip()
    p = k.find('div', class_='_8VNy32')

    pr = p.find('div', class_='_30jeq3').text.strip()
    new_price.append(pr)
    
    opr = p.find('div', class_='_3I9_wc')
    op = opr.text.strip() if opr else k.find('div', class_='_30jeq3').text.strip()
    original_price.append(op)
    
    dis_per = p.find('div', class_='_3Ay6Sb')
    dp = dis_per.text.strip() if dis_per else "No discount"
    discount_percentage.append(dp)
    
    
```

```python
s1 = pd.Series(model, name="model")
s2 = pd.Series(picture, name="picture")
s3 = pd.Series(URL, name="url")
s4 = pd.Series(new_price, name="new_price")
s5 = pd.Series(original_price, name="original_price")
s6 = pd.Series(discount_percentage, name="discount_percentage")
```

```python
df = pd.DataFrame({
    "model" : s1,
    "picture":s2,
    "url":s3, 
    "new_price":s4, 
    "original_price":s5, 
    "discounted_percentage":s6
})
```
