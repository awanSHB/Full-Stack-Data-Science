```python
df = pd.read_csv("D:\PROGRAMMING\DATASETS\ChicagoPublicSchools.csv")
print(df.shape)
df.head()
```

```python
chunksize = 100  # You can adjust this based on your data size
for i in range(0, len(df), chunksize):
    df.iloc[i:i+chunksize].to_sql("Chicago_School_Data", conn, if_exists='append', index=False)
```

* How many `Elementary Schools` are these in the dataset?

```python
%%sql
SELECT COUNT(*) as "Total_Elementary_Schools" 
FROM Chicago_School_Data 
WHERE "Elementary, Middle, or High School" = 'ES';
```

* Counting the `Frequency` of Schools

```python
%%sql

SELECT "Elementary, Middle, or High School" as "School_Names", COUNT(*) as "Frequency"
FROM Chicago_School_Data
GROUP BY "Elementary, Middle, or High School"
ORDER BY Frequency;
```

* What are `Top 10` schools with `highest` Average Student Attendance?

```python
%%sql 
SELECT "name_of_school", "average_student_attendance" 
from Chicago_School_Data
ORDER BY "average_student_attendance" DESC
LIMIT 10;
```

* What are `Top 10` schools with `lowest` Average Student Attendance?

```python
%%sql
SELECT "name_of_school" as "School", "average_student_attendance" as "average_attendance"
FROM Chicago_School_Data
ORDER BY "average_attendance"
LIMIT 10;
```

```python
%%sql

SELECT "name_of_school" as "School", REPLACE("average_student_attendance", '%', '') as "average_attendance"
FROM Chicago_School_Data
ORDER BY "average_attendance"
LIMIT 10;
```

```python
%%sql

SELECT "name_of_school" as "School", "average_student_attendance" as "average_attendance"
FROM Chicago_School_Data
WHERE "average_student_attendance" <= 70;
```
